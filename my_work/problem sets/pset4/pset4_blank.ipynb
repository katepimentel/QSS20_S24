```python
q1 = "SELECT COUNT(*) as row_count FROM math_gencompare"
row_count_df = pd.read_sql_query(q1, cnx)
row_count = row_count_df['row_count'].iloc[0]
print("Number of rows in the database:", row_count)

q2 = "SELECT COUNT(DISTINCT state) as distinct_states FROM math_gencompare"
distinct_states_df = pd.read_sql_query(q2, cnx)
distinct_states = distinct_states_df['distinct_states'].iloc[0]
print("Number of distinct states:", distinct_states)

q3 = "SELECT COUNT(DISTINCT year) as distinct_years FROM math_gencompare"
distinct_years_df = pd.read_sql_query(q3, cnx)
distinct_years = distinct_years_df['distinct_years'].iloc[0]
print("Number of distinct years:", distinct_years)
```

## 2.4 Construct a new variable, `is_male_higher` that takes the value of 1 if the math scores of males exceed that of females in that state and year (each row) (2 points)

Read in the results, print the head, and find the mean across all rows (the percentage of state-years where male students have higher scores)

```python
# Assuming you have already connected to your SQL database

# Constructing the new variable is_male_higher and retrieving the data
%%sql
SELECT 
    state,
    year,
    male_math_score,
    female_math_score,
    CASE WHEN male_math_score > female_math_score THEN 1 ELSE 0 END AS is_male_higher
FROM 
    math_gencompare;

# Finding the mean across all rows for is_male_higher
mean_percentage = %sql SELECT AVG(is_male_higher) FROM (SELECT CASE WHEN male_math_score > female_math_score THEN 1 ELSE 0 END AS is_male_higher FROM scores) AS subquery;
mean_percentage
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the gender disparity for 2013 (Male 2013 math scores - Female 2013 math scores)
acs_wmath_df['gender_disparity_2013'] = acs_wmath_df['math_male_2013'] - acs_wmath_df['math_female_2013']

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='abbrev', y='gender_disparity_2013', data=acs_wmath_df, palette='viridis')
plt.xticks(rotation=90)
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Gender Disparity in 2013 Math Scores by State')
plt.xlabel('State')
plt.ylabel('Male - Female Math Scores (2013)')
plt.tight_layout()

# Show the plot
plt.show()
```

## 3.2 Examine gender disparity in relation to household income (6 points)

**A.** Construct an indicator variable for the state having better performance of males in 2013 than females

**B.** First plot a scatterplot (or seaborn regplot) of estimated median household income from the acs data (we provide varname below) vs `math_male_2013`. Then do a second smoothed scatterplot for median household income vs `math_female_2013`.

Documentation: https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html

**D.** Interpret the correlations - in states with higher median household income (MHI), do 
   - boys tend to perform better than boys in states with lower MHI?
   - girls tend to perform better than girls in states with lower MHI?

```python
# PART A - Construct an indicator variable
acs_wmath_df['better_male_performance_2013'] = (acs_wmath_df['math_male_2013'] > acs_wmath_df['math_female_2013']).astype(int)
acs_wmath_df['better_male_performance_2013'].head()
```

```python
# B. Scatterplots of Median Household Income vs. Math Scores
median_income_var = 'acspredict_median_household_income_in_the_past_12_months__in_2018_inflation-adjusted_dollars_estimatemedian household income in the past 12 months in 2018 inflationadjusted dollars'

# Scatterplot for male math scores in 2013
plt.figure(figsize=(12, 6))
sns.regplot(x=median_income_var, y='math_male_2013', data=acs_wmath_df)
plt.title('Median Household Income vs. Male Math Scores (2013)')
plt.xlabel('Median Household Income (2018 inflation-adjusted dollars)')
plt.ylabel('Male Math Scores (2013)')
plt.show()

# Scatterplot for female math scores in 2013
plt.figure(figsize=(12, 6))
sns.regplot(x=median_income_var, y='math_female_2013', data=acs_wmath_df)
plt.title('Median Household Income vs. Female Math Scores (2013)')
plt.xlabel('Median Household Income (2018 inflation-adjusted dollars)')
plt.ylabel('Female Math Scores (2013)')
plt.show()
```

```python
# C. Calculate Correlations
# Correlation between male math scores and median household income
corr_male_income = np.corrcoef(acs_wmath_df['math_male_2013'], acs_wmath_df[median_income_var])[0, 1]

# Correlation between female math scores and median household income
corr_female_income = np.corrcoef(acs_wmath_df['math_female_2013'], acs_wmath_df[median_income_var])[0, 1]

# Correlation between indicator variable and median household income
corr_indicator_income = np.corrcoef(acs_wmath_df['better_male_performance_2013'], acs_wmath_df[median_income_var])[0, 1]

print(f"Correlation between male math scores and median household income: {corr_male_income}")
print(f"Correlation between female math scores and median household income: {corr_female_income}")
print(f"Correlation between indicator variable and median household income: {corr_indicator_income}")
```

In states with higher median household income (MHI), do boys tend to perform better than boys in states with lower MHI? 
The positive correlation (0.507) indicates that boys in states with higher median household income tend to have higher math scores. This suggests that economic factors play a significant role in boys' academic performance in math.

In states with higher median household income (MHI), do girls tend to perform better than girls in states with lower MHI? 
The stronger positive correlation (0.543) for girls suggests that girls' math performance is also positively influenced by higher median household income. The fact that this correlation is slightly higher than that for boys indicates that girls' math performance might be even more sensitive to economic factors.

In states with higher median household income (MHI), do boys tend to outperform girls more than they do in states with lower MHI?
The correlation between the indicator variable (which flags better male performance) and median household income is almost zero (-0.034). This suggests that economic factors, as measured by median household income, do not significantly affect the likelihood of boys outperforming girls in math. In other words, the gender disparity in math performance does not vary systematically with household income levels.
